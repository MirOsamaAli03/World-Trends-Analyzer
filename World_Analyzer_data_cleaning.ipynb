```python
df1=pd.read_csv("immunisation.csv")
df2=pd.read_csv("Life_Expectancy.csv")
df3=pd.read_csv("literacy_Rate.csv")
df4=pd.read_csv("Renewable_energy_consumption.csv")
df5=pd.read_csv("GDP1.csv")
df6=pd.read_csv("CO-Emissions.csv")
df7=pd.read_csv("child-mortality-rate.csv")

```

```python
cols_to_drop = ['Code']  # add any other columns you don't need

dfs = [df.drop(columns=cols_to_drop, errors='ignore') for df in [df1, df2, df3, df4, df5, df6, df7]]
```

```python
import dask.dataframe as dd

ddfs = [dd.from_pandas(df, npartitions=4) for df in dfs]
```

```python
merged = ddfs[0]
for ddf in ddfs[1:]:
    merged = dd.merge(merged, ddf, on=['Country','Year'], how='inner')  # or 'outer'
```

```python
merged_df = merged.compute()
```
